```python
# hapus node dengan kata kunci

class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None

class LinkedList:
    def __init__(self):
        self.headval = None

    # Print the linked list
    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval)
            printval = printval.nextval

    # function to add node in the beginning
    def AddBegin(self, newdata):
        NewNode = Node(newdata)
        NewNode.nextval = self.headval
        self.headval = NewNode

    # function to remove node by key
    def RemoveNodeByKey(self, key):
        HeadVal = self.headval

        # Jika node pertama adalah yang dicari
        if HeadVal is not None:
            if HeadVal.dataval == key:
                self.headval = HeadVal.nextval
                HeadVal = None
                return

        # Cari node dengan nilai yang sesuai
        prev = None
        while HeadVal is not None:
            if HeadVal.dataval == key:
                break
            prev = HeadVal
            HeadVal = HeadVal.nextval

        # Jika tidak ditemukan
        if HeadVal is None:
            return

        # Hapus node dengan menghubungkan prev ke next
        prev.nextval = HeadVal.nextval
        HeadVal = None


# create linked list named Li
Li = LinkedList()

# add node
Li.AddBegin("April")
Li.AddBegin("Maret")
Li.AddBegin("Februari")

print("Sebelum dilakukan Remove")
Li.listprint()

print("\n====================================")
Li.RemoveNodeByKey("Maret")

print("\nSetelah dilakukan remove")
Li.listprint()
```

```python
# 1. Buatlah double linked list dengan 3 node didalamnya, kemudian lakukan penyisipan tengah dan hapus dengan kata kunci !

class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.prev = None
        self.next = None

class DoubleLinkedList:
    def __init__(self):
        self.headval = None

    # Print the linked list
    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval)
            printval = printval.next

    # Function to add node in the middle
    def AddInbetween(self, mid_node, newdata):
        if mid_node is None:
            print("The mentioned node is absent")
            return
        NewNode = Node(newdata)
        NewNode.next = mid_node.next
        NewNode.prev = mid_node
        if mid_node.next is not None:
            mid_node.next.prev = NewNode
        mid_node.next = NewNode

    # Function to remove node by key
    def RemoveNode(self, Removekey):
        HeadVal = self.headval
        while HeadVal is not None:
            if HeadVal.dataval == Removekey:
                if HeadVal.prev is not None:
                    HeadVal.prev.next = HeadVal.next
                if HeadVal.next is not None:
                    HeadVal.next.prev = HeadVal.prev
                if HeadVal == self.headval:
                    self.headval = HeadVal.next
                HeadVal = None
                return
            HeadVal = HeadVal.next

# Membuat double linked list
dll = DoubleLinkedList()
dll.headval = Node("Apel")
second = Node("Mangga")
third = Node("Leci")

dll.headval.next = second
second.prev = dll.headval
second.next = third
third.prev = second

print("Sebelum penyisipan:")
dll.listprint()

dll.AddInbetween(dll.headval, "Semangka")
print("\nSetelah menyisipkan semangka ditengah:")
dll.listprint()

dll.RemoveNode("Leci")
print("\nSetelah menghapus dengan kata kunci 'Leci':")
dll.listprint()
```

```python
# 2. Buatlah circular linked list dengan 3 node didalamnya, kemudian lakukan penyisipan akhir dan hapus akhir !

class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.headval = None

    # Print the linked list
    def listprint(self):
        temp = self.headval
        if temp is None:
            return
        while True:
            print(temp.dataval)
            temp = temp.next
            if temp == self.headval:
                break

    # Function to add node in the end
    def AddInEnd(self, newdata):
        NewNode = Node(newdata)
        if self.headval is None:
            self.headval = NewNode
            NewNode.next = self.headval
            return
        temp = self.headval
        while temp.next != self.headval:
            temp = temp.next
        temp.next = NewNode
        NewNode.next = self.headval

    # Function to remove last node
    def RemoveEnd(self):
        if self.headval is None:
            return
        if self.headval.next == self.headval:
            self.headval = None
            return
        temp = self.headval
        prev = None
        while temp.next != self.headval:
            prev = temp
            temp = temp.next
        prev.next = self.headval
        temp = None

# Membuat circular linked list
cll = CircularLinkedList()
cll.AddInEnd("Apel")
cll.AddInEnd("Mangga")
cll.AddInEnd("Jeruk")

print("Sebelum penyisipan:")
cll.listprint()

cll.AddInEnd("Semangka")
print("\nSetelah penyisipan akhir (Semangka):")
cll.listprint()

cll.RemoveEnd()
print("\nSetelah hapus akhir:")
cll.listprint()
```
