```python
# Mwmbuat Binary Tree Secara Manual

class Node:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

def inorder(node):
  if node:
    inorder(node.left)
    print(node.data, end='')
    inorder(node.right)

print("Hasil inorder traversal dari binary tree")
inorder(root)
```

```python
# Membuat Binary Search Tree

class Node:
  def __init__(self,data):
    self.data = data
    self.left = None
    self.right = None

class BinarySearchTre:
  def __init__ (self):
    self.root = None
  def insert(self, root, data):
    if root is None:
      return Node(data)
    if data < root.data:
      root.left = self.insert(root.left, data)
    else:
      root.right = self.insert(root.right, data)
    return root

bst = BinarySearchTre()
root = None
data_list = [50, 30, 70, 20, 40, 60, 80]
for value in data_list:
  root = bst.insert(root, value)
```

```python
# Traversal pada BST

def inorder(node):
  if node:
    inorder(node.left)
    print(node.data, end='')
    inorder(node.right)
def preorder(node):
  if node:
    print(node.data, end='')
    preorder(node.left)
    preorder(node.right)
def postorder (node):
  if node:
    postorder(node.left)
    postorder(node.right)
    print(node.data, end='')

print("Hasil inorder traversal dari binary tree")
inorder(root)
print("\nHasil preorder traversal dari binary tree")
preorder(root)
print("\nHasil postorder traversal dari binary tree")
postorder(root)
```

```python
# Pencarian Nilai dalam BST

def search(node,key):
  if node is None or node.data ==key:
    return node
  if key < node.data:
    return search(node.left, key)
  return search(node.right, key)

key = 60
result = search(root, key)
if result:
  print(f"Node dengan nilai {key} ditemukan")
else:
  print(f"Node dengan nilai {key} tidak ditemukan")

key = 25
result = search(root, key)
if result:
  print(f"Node dengan nilai {key} ditemukan")
else:
  print(f"Node dengan nilai {key} tidak ditemukan")
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
class BST:
    def __init__(self):
        self.root = None

    # Fungsi insert
    def insert(self, data):
        if self.root is None:
            self.root = Node(data)
        else:
            self._insert(self.root, data)

    def _insert(self, node, data):
        if data < node.data:
            if node.left is None:
                node.left = Node(data)
            else:
                self._insert(node.left, data)
        elif data > node.data:
            if node.right is None:
                node.right = Node(data)
            else:
                self._insert(node.right, data)

    # Inorder Traversal (Left, Root, Right)
    def inorder(self, node):
        if node:
            self.inorder(node.left)
            print(node.data, end=" ")
            self.inorder(node.right)

    # Preorder Traversal (Root, Left, Right)
    def preorder(self, node):
        if node:
            print(node.data, end=" ")
            self.preorder(node.left)
            self.preorder(node.right)

    # Postorder Traversal (Left, Right, Root)
    def postorder(self, node):
        if node:
            self.postorder(node.left)
            self.postorder(node.right)
            print(node.data, end=" ")

    # Fungsi pencarian
    def search(self, node, key):
        if node is None:
            return False
        if key == node.data:
            return True
        elif key < node.data:
            return self.search(node.left, key)
        else:
            return self.search(node.right, key)

nama = "Pangestu Ilham Dwi Prastyo"
npm = "2505060026"
tanggal_lahir = "07-11-2005"

#Tiga digit terakhir NPM
tiga_digit_npm = int(npm[-3:])  # 022 → 22

#Tanggal dan bulan
tanggal = int(tanggal_lahir.split("-")[0])
bulan = int(tanggal_lahir.split("-")[1])

#ASCII huruf pertama dan kedua nama depan
ascii1 = ord(nama[0])
ascii2 = ord(nama[1])

#Dua angka bebas
bebas1 = 8
bebas2 = 11

# Daftar angka unik
data_bst = list(set([
    tiga_digit_npm,
    tanggal,
    bulan,
    ascii1,
    ascii2,
    bebas1,
    bebas2
]))

print("Data yang dimasukkan ke BST:", data_bst)

tree = BST()

for angka in data_bst:
    tree.insert(angka)

print("\nInorder Traversal:")
tree.inorder(tree.root)

print("\n\nPreorder Traversal:")
tree.preorder(tree.root)

print("\n\nPostorder Traversal:")
tree.postorder(tree.root)

# Cari dua digit terakhir NPM
cari1 = int(npm[-2:])
print("\n\nMencari:", cari1)
print("Ditemukan?" , tree.search(tree.root, cari1))

# Cari angka yang tidak ada
cari2 = 2
print("Mencari:", cari2)
print("Ditemukan?" , tree.search(tree.root, cari2))
```
